```python
import tkinter as tk
from tkinter import filedialog
from tkinter.ttk import Combobox
import pyttsx3
import os
```

```python
# Initialize the text-to-speech engine
tts = pyttsx3.init()

root = tk.Tk()
root.title("Simplilearn Text to Speech Converter")

# Window size and positioning (Width x Height + X_offset + Y_offset)
root.geometry("1200x780+250+100")
root.resizable(False, False)
root.configure(bg="#f7ac40")
```

```python
# Header Frame
upper_frame = tk.Frame(root, bg="#14a7dd", width=1200, height=130)
upper_frame.place(x=0, y=0)

# Application Title Label inside the top frame
heading = tk.Label(upper_frame, text="TEXT TO SPEECH CONVERTER", font=("Times New Roman", 40, "bold"), bg="#14a7dd", fg="white")
heading.place(x=250, y=35)
```

```python
# Input text area
text_box = tk.Text(root, font="Roboto 20", bg="white", relief="groove", wrap="word", bd=0)
text_box.place(x=30, y=150, width=940, height=180)
```

```python
# Option Labels
tk.Label(root, text="SELECT VOICE TYPE", font="times 15 bold", bg="#f7ac40", fg="white").place(x=40, y=370)
tk.Label(root, text="SELECT SPEED", font="times 15 bold", bg="#f7ac40", fg="white").place(x=540, y=370)

# Dropdown for selecting Gender
gender_box = Combobox(root, values=["Male", "Female"], font="Roboto 12", state="readonly", width=12)
gender_box.place(x=40, y=400)
gender_box.set("Male")

# Dropdown for selecting Speed
speed_box = Combobox(root, values=["Fast", "Medium", "Slow"], font="Roboto 12", state="readonly", width=12)
speed_box.place(x=540, y=400)
speed_box.set("Medium")
```

```python
def speak_now():
    text = text_box.get(1.0, tk.END).strip()
    gender = gender_box.get()
    speed = speed_box.get()
    
    voices = tts.getProperty('voices')
    
    # Internal function to switch voice gender profiles
    def set_voice():
        if gender == "Male":
            tts.setProperty('voice', voices[0].id)
        else:
            tts.setProperty('voice', voices[1].id)
            
    if text:
        # Set speaking rate speed configuration
        if speed == "Fast":
            tts.setProperty('rate', 250)
        elif speed == "Medium":
            tts.setProperty('rate', 150)
        else:
            tts.setProperty('rate', 60)
            
        set_voice()
        tts.say(text)
        tts.runAndWait()

# Interactive Play Button widget linked to backend voice generator
play_btn = tk.Button(root, text="Play", bg="white", width=13, font="Arial 14 bold", command=speak_now)
play_btn.place(x=435, y=450)
```

```python
root.mainloop()
```
